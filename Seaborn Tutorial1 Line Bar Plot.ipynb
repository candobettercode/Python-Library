```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Line Plot

```python
var = [1,2,3,4,5,6,7]
var1 = [2,3,4,5,1,6,7]
```

```python
plt.plot(var,var1)
plt.show()
```

```python
x = pd.DataFrame({'var':var, 'var1':var1})
```

```python
sns.lineplot(x='var',y='var1',data = x)
plt.show()
```

```python
sns.lineplot(x=var,y=var1)
plt.show()
```

```python
peg = sns.load_dataset('penguins')
```

```python
peg = peg.head(40)
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex',size=70)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex',style='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex',style='sex',palette='plasma')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex',style='sex',palette='plasma',markers=['o','>'])
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex',style='sex',palette='plasma',markers=['o','>'],dashes=False)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data =peg,hue='sex',
             style='sex',palette='plasma',markers=['o','>'],
             dashes=False,legend=False)
plt.show()
```

## Bar Plot

```python
var = sns.load_dataset('penguins')
```

```python
var.head()
```

```python
sns.barplot(x=var.island, y = var.bill_length_mm)
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,hue = 'sex')
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y = 'bill_depth_mm', data = var,hue = 'sex',
           order=order1)
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
            hue = 'sex', order=order1, hue_order=['Female','Male']
           )
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
            hue = 'sex', order=order1, hue_order=['Female','Male'],
           ci=100)
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
            hue = 'sex', order=order1, hue_order=['Female','Male'],
           ci=100,n_boot=10)
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           color='b')
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex')
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=0.3)
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=0.3,
           errcolor='g')
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=0.3,
           errcolor='g',errwidth=12)
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=0.3,
           errcolor='g',capsize=0.3)
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=0.3,
           errcolor='g',dodge=False)
plt.show()
```

```python
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=1,
           errcolor='g',alpha = 0.5)
plt.show()
```

```python
sns.set(style='')
sns.barplot(x='island', y = 'bill_depth_mm', data = var,
           palette='icefire',hue='sex',saturation=1,
           errcolor='g',alpha = 0.5)
plt.show()
```
