```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Violin Plot 

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
              data = var)
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
              data = var,
            hue ='time')
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
              data = var,
            hue ='time',
              linewidth=2)
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
              data = var,
              hue ='time',
              linewidth=2,
              palette='dark')
plt.show()
```

```python
sns.violinplot(x='time',
              y = 'total_bill',
              data = var,
              order=['Dinner','Lunch'])
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
              data = var,
              hue ='time',
              linewidth=2,
              palette='dark',
              saturation=0.1)
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
              data = var,
              hue ='time',
              linewidth=2,
              color='y')
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
               hue = 'sex',
              data = var,
              split=True)
plt.show()
```

```python
sns.violinplot(x='day',
              y = 'total_bill',
               hue = 'sex',
              data = var,
              split=True,
              scale='count')
plt.show()
```

```python
sns.violinplot(x = 'total_bill',
               y='day',
              hue = 'sex',
              data = var,
              split=True,
              scale='count')
plt.show()
```

```python
sns.violinplot(x='time',
              y = 'total_bill',
              data = var,
              order=['Dinner','Lunch'],
              inner='quart')
plt.show()
```

```python
sns.violinplot(x='time',
              y = 'total_bill',
              data = var,
              order=['Dinner','Lunch'],
              inner='stick')
plt.show()
```

```python
sns.violinplot(x=var['total_bill'])
plt.show()
```

```python
sns.violinplot(y=var['total_bill'])
plt.show()
```

## Pair Plot

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,
            hue='sex')
plt.show()
```

```python
sns.pairplot(var,
             vars=['total_bill','tip'],
            hue='sex')
plt.show()
```

```python
sns.pairplot(var,
             vars=['total_bill','tip'],
            hue='sex',
            hue_order=['Female','Male'])
plt.show()
```

```python
sns.pairplot(var,
             vars=['total_bill','tip'],
            hue='sex',
            hue_order=['Female','Male'],
            palette='spring')
plt.show()
```

```python
sns.pairplot(var,
            hue='sex',
            hue_order=['Female','Male'],
            x_vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(var,
            hue='sex',
            hue_order=['Female','Male'],
            y_vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(var,
            hue='sex',
            hue_order=['Female','Male'],
            kind='reg')#scatter,kde,hist,reg
plt.show()
```

```python
sns.pairplot(var,
            hue='sex',
            hue_order=['Female','Male'],
            kind='kde')#scatter,kde,hist,reg
plt.show()
```

```python
sns.pairplot(var,
            hue='sex',
            hue_order=['Female','Male'],
            kind='kde',
            diag_kind='hist')#scatter,kde,hist,reg
plt.show()
```

```python
sns.pairplot(var,
            hue='sex',
            hue_order=['Female','Male'],
            markers=['*','v'])
plt.show()
```
