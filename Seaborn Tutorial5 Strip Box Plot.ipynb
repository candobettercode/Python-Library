```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Strip Plot

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.stripplot(x='day',y='total_bill',data= var)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex',
             palette='rocket')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex',
             palette='rocket',
             linewidth=0.5)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex',
             palette='rocket',
             linewidth=0.5,
             edgecolor='b')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex',
             palette='rocket',
             linewidth=0.5,
             edgecolor='b',
             jitter=1)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex',
             palette='rocket',
             linewidth=0.5,
             edgecolor='b',
             size=10)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',
              data= var,
             hue='sex',
             palette='rocket',
             linewidth=0.5,
             edgecolor='b',
             alpha =0.6)
plt.show()
```

```python
sns.stripplot(x='total_bill',data = var)
plt.show()
```

```python
sns.stripplot(y='total_bill',data = var)
plt.show()
```

## Box Plot

```python
box = sns.load_dataset('tips')
```

```python
box
```

```python
sns.set(style='whitegrid')
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box)
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'time')
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex')
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y')
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y',
           order=['Fri','Sun','Thur','Sat'])
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y',
           order=['Fri','Sun','Thur','Sat'],
           showmeans = True)
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y',
           order=['Fri','Sun','Thur','Sat'],
           showmeans = True,
           meanprops = {'marker':'+','markeredgecolor':'b'})
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y',
           order=['Fri','Sun','Thur','Sat'],
           showmeans = True,
           meanprops = {'marker':'+','markeredgecolor':'b'},
           linewidth=3)
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y',
           order=['Fri','Sun','Thur','Sat'],
           showmeans = True,
           meanprops = {'marker':'+','markeredgecolor':'b'},
           linewidth=3,
           palette='plasma')
plt.show()
```

```python
sns.boxplot(x = 'day', y='total_bill',
           data = box,
           hue = 'sex',
           color='y',
           order=['Fri','Sun','Thur','Sat'],
           showmeans = True,
           meanprops = {'marker':'+','markeredgecolor':'b'},
           linewidth=3,
           palette='plasma',
           orient='v')
plt.show()
```

```python
sns.boxplot(x='total_bill',y = 'day', 
           data = box,
           order=['Fri','Sun','Thur','Sat'],
           showmeans = True,
           meanprops = {'marker':'+','markeredgecolor':'b'},
           linewidth=3,
           palette='plasma')
plt.show()
```

```python
sns.boxplot(data = box,
           showmeans = True,
           meanprops = {'marker':'+','markeredgecolor':'b'},
           linewidth=3,
           palette='plasma',
           orient='h')
plt.show()
```
