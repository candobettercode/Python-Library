```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Histogram

```python
var = sns.load_dataset('penguins')
```

```python
var
```

```python
sns.displot(var.bill_length_mm)
plt.show()
```

```python
sns.displot(var.flipper_length_mm,bins =[170,190,210,230,250])
plt.show()
```

```python
sns.displot(var.flipper_length_mm,
            bins =[170,190,210,230,250],
           kde=True)
plt.show()
```

```python
sns.displot(var.flipper_length_mm,
            bins =[170,190,210,230,250],
           kde=True,
           rug=True)
plt.show()
```

```python
sns.displot(var.flipper_length_mm,
            bins =[170,190,210,230,250],
           kde=True,
           rug=True,
           color='b')
plt.show()
```

```python
sns.displot(var.flipper_length_mm,
           kde=True,
           rug=True,
           color='b',
           log_scale=True)
plt.show()
```

## Scatter Plot
