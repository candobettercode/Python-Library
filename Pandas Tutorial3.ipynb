### Passing data to Functions

```python
import sys
```

```python
df = pd.read_csv('ComputerSales.csv',index_col=0)
```

```python
df
```

```python
def calculate_profit(df):
    profit = df['Profit']
    print(f"Total Profit: {profit.sum()}")
```

```python
calculate_profit(df)
```

```python
def split_name(df):
    def get_names(full_name):
        f_name, l_name = full_name.split()
        return pd.Series((f_name,l_name), index = ['First Name','Last Name'])
    names = df['Contact'].apply(get_names)
    df[names.columns] = names
    return df
```

```python
split_name(df).head()
```

```python
def create_age_group(df):
    bins = [0,30,50,sys.maxsize]
    lables = ['<30','30-50','>50']
    age_group = pd.cut(df['Age'],bins=bins, labels=lables)
    df['Age Gorup'] = age_group
    return df
```

```python
create_age_group(df)
```
