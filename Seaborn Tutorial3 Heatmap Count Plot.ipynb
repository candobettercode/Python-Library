```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Heatmap

```python
var = np.linspace(1,10,20).reshape(4,5)
var
```

```python
sns.heatmap(var)
plt.show()
```

```python
datas = sns.load_dataset('anagrams')
```

```python
datas = datas.drop(columns=['attnr'],axis=1)
```

```python
datas = datas.head(10)
datas
```

```python
sns.heatmap(datas)
plt.show()
```

```python
sns.heatmap(var,
           vmin = 0,
           vmax= 15)
plt.show()
```

```python
sns.heatmap(var,
           vmin = 0,
           vmax= 10,
           cmap='PuOr')
plt.show()
```

```python
sns.heatmap(var,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True)
plt.show()
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True)
plt.show()
```

```python
a = {'fontsize':20,'color':'r'}
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           annot_kws=a)
plt.show()
```

```python
ar = np.array([['a0','a1','a2','a3','a4'],
               ['b0','b1','b2','b3','b4']],dtype ='<U2')
```

```python
ar
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=ar,fmt='s')
plt.show()
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           linewidths=5)
plt.show()
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           linewidths=5,
           linecolor='y')
plt.show()
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           linewidths=5,
           linecolor='y',
           cbar=False)
plt.show()
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           linewidths=5,
           linecolor='y',
           cbar=True,
           xticklabels=False)
plt.show()
```

```python
sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           linewidths=5,
           linecolor='y',
           cbar=True,
           xticklabels=False,
           yticklabels=False)
plt.show()
```

```python
v = sns.heatmap(var1,
           vmin = 0,
           vmax= 10,
           cmap='PuOr',
           annot=True,
           linewidths=5,
           linecolor='y',
           cbar=True,
           xticklabels=False,
           yticklabels=False)
v.set(xlabel = 'python',ylabel = 'java')
sns.set(font_scale=1)
plt.show()
```

## Count Plot

```python
var = sns.load_dataset('tips')
```

```python
var
```

```python
sns.countplot(x='sex',data=var)
plt.show()
```

```python
sns.barplot(x='sex',y='size',data = var)
plt.show()
```

```python
sns.countplot(x='sex',data=var, hue='smoker')
plt.show()
```

```python
sns.countplot(y='sex',data=var, hue='smoker')
plt.show()
```

```python
sns.countplot(x='sex',
              data=var,
              hue='smoker',
              palette='hot')
plt.show()
```

```python
sns.countplot(x='sex',
              data=var,
              hue='smoker',
              color='y')
plt.show()
```

```python
sns.countplot(x='sex',
              data=var,
              hue='smoker',
              color='y',
              saturation=0.2)
plt.show()
```
