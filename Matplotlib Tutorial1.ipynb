```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
```

```python
#plt.plot(x,y, color ='red',label = '2X',linestyle = '--', linewidth =5, marker = 'o',markersize = 10,markeredgecolor = 'blue')

# Resize graph in inches
plt.figure(figsize=(6,4),dpi = 300)

plt.title('First Grpah',fontdict={'fontname':'Comic Sans Ms','fontsize':20})

# short hand notation
#[color][marker][line]
#plt.plot(x,y,'r.-',label ='2X') #solid line
#plt.plot(x,y,'r.--',label ='2X') # dash line
plt.plot(x,y,'g^-',label ='2X')

# line no 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'ro-',label = 'x^2')
plt.plot(x2[5:],x2[5:]**2,'ro--')

plt.xlabel('X Axis',fontdict={'fontname':'Arial'})
plt.ylabel('Y Axis')

#changing the numbers on axis (scale of the graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,7,7.5,8,10])

plt.legend()

plt.savefig('mygraph.png',dpi=600)

plt.show()
```
