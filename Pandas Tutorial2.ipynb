```python
dict_5 = {'Store':[1,2,1,2],
          'Flavor': ['Choc','Van','Straw','Choc'],'Sales': [26,12,18,22]}
```

```python
df_2 = pd.DataFrame(dict_5)
```

```python
df_2
```

```python
by_store = df_2.groupby('Store')
```

```python
print(by_store)
```
